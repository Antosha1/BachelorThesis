```python
import urllib
import requests
from tqdm import tqdm
```

```python
f = open('../corpus.txt')

ids = []

for line in f:
    candidate = line.split()[0]
    if (candidate.isdigit()):
        ids.append(candidate)
```

```python
ids[:10]
```

```python
sent_annot = {}

pbar = tqdm(ids)

for id in pbar:
    path = 'https://raw.githubusercontent.com/rockt/SETH/master/resources/SETH-corpus/annotations/' + str(id) + '.ann'
    annot = urllib.request.urlopen(path).read()
    sent_annot[id] = annot
```

```python
for item in sent_annot.items():
        print(item)
```

```python
import pickle

with open('annotations.pickle', 'wb') as f:
    pickle.dump(sent_annot, f)
```
